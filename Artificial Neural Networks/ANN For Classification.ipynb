```python
X=pd.get_dummies(df.drop('Exited',axis=1),drop_first=True)
y=df['Exited']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,
                                                 random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

### Modelling 
#### Initializing the ANN

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
### Initialising ANN
ann_model=Sequential()

### Adding the input layer and first hidden layer
ann_model.add(Dense(input_dim=11,units=21,activation='relu',
                    kernel_initializer='uniform'))

### Adding the second hidden layer
ann_model.add(Dense(units=21,activation='relu',
                    kernel_initializer='uniform'))
## No.of hidden layers=2(inputlayers)--->Max, 
## Where overfitting is not coming then that value is minimum

### Adding the output layer
ann_model.add(Dense(units=1,activation='sigmoid',
                    kernel_initializer='uniform'))

### Compiling the ANN
ann_model.compile(optimizer='adam',loss='binary_crossentropy',
                 metrics=['accuracy'])
```

```python
##### For multi-class(Penguin dataset)

##units of output layer=no.of categories in output variable Here 3**

##Activation function of output layer = 'softmax'

# Loss= "categorical_crossentropy"
```

#### Training the ANN on the Training Set

```python
ann_model.fit(X_train,y_train,batch_size=32,epochs=100)
```

#### Predictions

```python
ypred_train=ann_model.predict(X_train)
ypred_train= (ypred_train>=0.5)

ypred_test=ann_model.predict(X_test)
ypred_test= (ypred_test>=0.5)
```

#### Evaluation

```python
from sklearn.metrics import confusion_matrix,accuracy_score

print('Test Accuracy:',accuracy_score(y_test,ypred_test))
print('Train Accuracy:',accuracy_score(y_train,ypred_train))
```

```python
confusion_matrix(y_test,ypred_test)
```

#### Cross validate the model

```python
def build_cross_classifier():
    
    classifier=Sequential()

    classifier.add(Dense(input_dim=11,units=10,activation='relu',
                        kernel_initializer='uniform'))
    classifier.add(Dense(units=10,activation='relu',
                        kernel_initializer='uniform'))
    classifier.add(Dense(units=1,activation='sigmoid',
                        kernel_initializer='uniform'))

    classifier.compile(optimizer='adam',loss='binary_crossentropy',
                     metrics=['accuracy'])
    return classifier
```

```python
from scikeras.wrappers import KerasClassifier
classifier=KerasClassifier(build_cross_classifier(),batch_size=32,
                           epochs=100)
```
